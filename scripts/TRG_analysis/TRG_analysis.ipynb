### Analyse the genes from line 49, 51, and 63

```python
# Keep only line 49, 51, 63
choice = data.loc[[47, 49, 61]].reset_index(drop=True)
print(f"Here are the TRGs we're going to look at for {GENOME}:")
choice.head()
```

```python
# Get the last gS for each row
last_gS = []
for idx in choice.index:
    last_gS_row = 1
    for col_idx in range(1, len(choice.columns)):
        col = choice.columns[col_idx]
        if choice.at[idx, col] == "gS":
            last_gS_row = col_idx
    last_gS.append(last_gS_row)

print("For each TRG, the last gS is at the row n°:")
print(last_gS)
```

```python
# Get the CDSs after the last gS for each row
CDSs = {}
for idx in choice.index:
    gene = choice.iloc[idx, 0]
    CDSs[gene] = []
    for col_idx in range(last_gS[idx] + 1, len(choice.columns)):
        col = choice.columns[col_idx]
        if choice.at[idx, col] == "CDS":
            CDSs[gene].append(col)

for gene, CDS in CDSs.items():
    print(f"TRG {gene} has {len(CDS)} CDSs after the last gS.")
    print("They are in the following genomes:")
    print(f"{CDS}\n")
```
